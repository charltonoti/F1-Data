```python
circuits = pd.read_csv('circuits.csv', usecols= ['name', 'location', 'lat', 'lng'])
races = pd.read_csv('races.csv', usecols= ['year','round','name'])

races

```

```python
my_map = folium.Map(
        location = [-37.84970, 144.96800], zoom_start = 2)
my_map
```

```python
#create a macro to pin point to a circuit

circuit = circuits.loc[0]
folium.Marker(
        location = [circuit['lat'], circuit['lng']]
).add_to(my_map)

my_map
```

```python
#create a macro to pin point to all circuits
#my_map = folium.Map(
        #location = [-37.84970, 144.96800], zoom_start = 2)
#my_map

#circuit = df.loc[20]
for _, circuit in circuits.iterrows(): 
    folium.Marker(
            location = [circuit['lat'], circuit['lng']], 
            popup=circuit['name'],
            tooltip=circuit['name']
    ).add_to(my_map)

my_map

```
